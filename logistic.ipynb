```python
#importing required libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
#reading the data from csv file and printing
df=pd.read_csv("iris2.csv",names=["sl","sw","pl","pw","l"])
df
```

```python
#for mean,median,std,min,quartiles,max
df.describe()
```

```python
#gives the information like type of data,null or not null,column headings
df.info()
```

```python
fet=df.iloc[:,:-1].values
label=df.iloc[:,-1].values
fet
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(fet,label,test_size=0.2)
```

```python
lg=LogisticRegression()
lg.fit(x_train,y_train)
```

```python
y_pred=lg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report

print(classification_report(y_test,y_pred))
```
